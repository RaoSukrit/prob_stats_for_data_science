```python
import matplotlib.pyplot as plt
import numpy as np
from os import listdir
from collections import Counter

font_size = 30
font_size_ticks = 25

np.set_printoptions(precision=5)

def process_name(x):
    x = x[14:]
    x = x[:-7]
    x = x.translate(str.maketrans('','','_1234567890'))
    return x[2:] + ", " + x[:2]

file_path = "./data/weather/hourly_precipitation_2015.npy"

# data_matrix contains precipitation data from 134 stations (each station is a column)
data_matrix = np.load(file_path)

print(data_matrix.shape) 
print(data_matrix[25:35,:10])
```

```python
# We select three stations to study, stations contains the corresponding indices in data_matrix
# 24: Bodega, CA
# 99: Coos Bay, OR
# 102: Riley, OR
stations = [24,99,102]
```

#### preprocessing steps. We remove rows with fault readings from all stations (if we remove row 2 from bodega, we must remove it from coos bay and riley as well


```python
def data_preprocessing(station_1,station_2,station_3,data_matrix):
    # index out the columns corresponding to stations of interest
    bodega_data = data_matrix[:, station_1]
    coos_bay_data = data_matrix[:, station_2]
    riley_data = data_matrix[:, station_3]  
    
#     bodega_data[bodega_data < 0.0] = 0.0
#     coos_bay_data[coos_bay_data < 0.0] = 0.0
#     riley_data[riley_data < 0.0] = 0.0    
        
    # find index values which have valid readings (specifically, we ignore values that are negative)
    clean_bodega_data_ind = np.where(bodega_data >= 0)[0]
    clean_coos_bay_data_ind = np.where(coos_bay_data >= 0)[0]
    clean_riley_data_ind = np.where(riley_data >= 0)[0]
    
    # take the intersection of valid readings for all stations. We need to have readings of the same time frm all stations otherwise the pmf won't make sense    
    combined_clean_data_ind = set(clean_bodega_data_ind).intersection(set(clean_coos_bay_data_ind)).intersection(set(clean_riley_data_ind))
    combined_clean_data_ind = np.array(list(combined_clean_data_ind))
    
    # index values that have valid (non-negative precipitation readings)
    clean_bodega_data = bodega_data[combined_clean_data_ind]
    clean_coos_bay_data = coos_bay_data[combined_clean_data_ind]
    clean_riley_data = riley_data[combined_clean_data_ind]
    
    # combine the readings from all stations into one matrix
    clean_data_combined = np.column_stack([clean_bodega_data, clean_coos_bay_data, clean_riley_data])
    
    # find indices that have non-zero precipitation readings and convert this to 1s. Keep the entries with 0 precipitation as 0
    preprocessed_data = np.where(clean_data_combined > 0, 1, 0)
        
    return preprocessed_data
```

```python
preprocessed_data = data_preprocessing(stations[0],stations[1],stations[2],data_matrix)
preprocessed_data.shape
```

```python
# Compute joint pmf of three Bernoulli random variables indicating whether it rains (1) or not (0)
# in Bodega, Coos Bay and Riley
def compute_joint_pmf(station_1,station_2,station_3,data_matrix):    
    preprocessed_data = data_preprocessing(stations[0],stations[1],stations[2],data_matrix)
    
    print(preprocessed_data.shape)
    
    row_wise_data_arr = np.split(preprocessed_data, preprocessed_data.shape[0], axis=0)
    row_wise_data_tup = list(map(lambda row: tuple(row.flatten()), row_wise_data_arr))

    occurrences = Counter(row_wise_data_tup)
    total_count = sum(occurrences.values())
    
    joint_pmf = np.zeros((2, 2, 2))
    counts_array = np.zeros((2, 2, 2))
    
    for key, count in occurrences.items():
        joint_pmf[key] = count / total_count
        counts_array[key] = count
        
    return counts_array, joint_pmf

counts,joint_pmf = compute_joint_pmf(stations[0],stations[1],stations[2],data_matrix)
print(counts)
print(joint_pmf)
```

```python
# Compute marginal pmf of each of the Bernoulli random variables 
def marginal_1_station(joint_pmf):
    # INSERT YOUR CODE HERE
    marginal_pmf_1 = joint_pmf.sum(axis=2).sum(axis=1)
    marginal_pmf_2 = joint_pmf.sum(axis=0).sum(axis=1)
    marginal_pmf_3 = joint_pmf.sum(axis=1).sum(axis=0)
    
    return marginal_pmf_1, marginal_pmf_2, marginal_pmf_3 

marginal_pmf_1,marginal_pmf_2,marginal_pmf_3 = marginal_1_station(joint_pmf)
print(marginal_pmf_1)
print(marginal_pmf_2)
print(marginal_pmf_3)

vals = [0,1]
ymax = 1.0
xmin = -0.6
xmax = 1.6
plt.figure(figsize=(6,9))
plt.bar(vals,marginal_pmf_1, width = 0.5, color = "lightgray", edgecolor="black",linewidth = 2)
plt.xticks(np.arange(0, 1+1, 1))
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylim([0,ymax])
plt.xlim([xmin,xmax])
plt.ylabel("Probability mass function",fontsize=font_size,labelpad = 30)
plt.xlabel("Precipitation in Bodega",fontsize=font_size,labelpad = 15)
# plt.savefig('plots/precipitation_marginal_pmf_1.pdf',bbox_inches="tight")

plt.figure(figsize=(6,9))
plt.bar(vals,marginal_pmf_2, width = 0.5, color = "lightgray", edgecolor="black",linewidth = 2)
plt.xticks(np.arange(0, 1+1, 1))
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylim([0,ymax])
plt.xlim([xmin,xmax])
plt.ylabel("Probability mass function",fontsize=font_size,labelpad = 30)
plt.xlabel("Precipitation in Coos Bay",fontsize=font_size,labelpad = 15)
# plt.savefig('plots/precipitation_marginal_pmf_2.pdf',bbox_inches="tight")

plt.figure(figsize=(6,9))
plt.bar(vals,marginal_pmf_3, width = 0.5, color = "lightgray", edgecolor="black",linewidth = 2)
plt.xticks(np.arange(0, 1+1, 1))
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylim([0,ymax])
plt.xlim([xmin,xmax])
plt.ylabel("Probability mass function",fontsize=font_size,labelpad = 30)
plt.xlabel("Precipitation in Riley",fontsize=font_size,labelpad = 15)
# plt.savefig('plots/precipitation_marginal_pmf_3.pdf',bbox_inches="tight")
```

```python
# Compute marginal joint pmf of each pair of the Bernoulli random variables 
def marginal_2_stations(joint_pmf):
    # INSERT YOUR CODE HERE
    marginal_pmf_12 = joint_pmf.sum(axis=2)
    marginal_pmf_13 = joint_pmf.sum(axis=1)
    marginal_pmf_23 = joint_pmf.sum(axis=0)

    return marginal_pmf_12,marginal_pmf_13,marginal_pmf_23

marginal_pmf_12,marginal_pmf_13,marginal_pmf_23 = marginal_2_stations(joint_pmf)

print(marginal_pmf_12)
print(marginal_pmf_13)
print(marginal_pmf_23)
```

```python
# Compute conditional pmf of each of the Bernoulli random variables given the other two 
def conditional_1_station_given_2(joint_pmf):
    # INSERT YOUR CODE HERE
    cond_1_given_23 = np.zeros((2, 2, 2))
    cond_2_given_13 = np.zeros((2, 2, 2))
    cond_3_given_12 = np.zeros((2, 2, 2))    
    
    for index1 in range(2):
        for index2 in range(2): 
            for index3 in range(2):
                cond_1_given_23[index1, index2, index3] = joint_pmf[index1, index2, index3] / joint_pmf[:, index2, index3].sum()
                cond_2_given_13[index2, index1, index3] = joint_pmf[index2, index1, index3] / joint_pmf[index2, :, index3].sum()
                cond_3_given_12[index2, index3, index1] = joint_pmf[index2, index3, index1] / joint_pmf[index2, index3, :].sum()

    return cond_1_given_23,cond_2_given_13,cond_3_given_12
```

```python
# Compute conditional pmf of each Bernoulli random variable given each of the other ones 
# (i.e. Bodega just conditioned on Coos Bay, Bodega just conditioned on Riley, etc.)
# Use a dictionary to save the conditional pmfs, for example cond_1["2"] should contain the conditional of the
# first random variable (Bodega)
def conditional_1_station_given_1_station(counts, joint_pmf):
    # INSERT YOUR CODE HERE
    cond_1 = {"2" : np.zeros((2, 2)), "3": np.zeros((2, 2))}
    cond_2 = {"1" : np.zeros((2, 2)), "3": np.zeros((2, 2))}    
    cond_3 = {"1" : np.zeros((2, 2)), "2": np.zeros((2, 2))}
    
    for index1 in range(2):
        for index2 in range(2):                
                cond_1["2"][index2, index1] = counts[index2, index1, :].sum() #/ joint_pmf[:, index1, :].sum()
                cond_1["3"][index2, index1] = counts[index2, :, index1].sum() #/ joint_pmf[:, :, index1].sum()                

    for index1 in range(2):
        for index2 in range(2):
                cond_2["1"][index2, index1] = counts[index1, index2, :].sum() #/ joint_pmf[index1, :, :].sum()
                cond_2["3"][index2, index1] = counts[:, index2, index1].sum() #/ joint_pmf[:, :, index1].sum()

    for index1 in range(2):
        for index2 in range(2):
                cond_3["1"][index2, index1] = counts[index1, :, index2].sum() #/ joint_pmf[index1, :, :].sum()
                cond_3["2"][index2, index1] = counts[:, index1, index2].sum() #/ joint_pmf[:, index2, :].sum()

                    
    return cond_1, cond_2, cond_3
    
    
```
